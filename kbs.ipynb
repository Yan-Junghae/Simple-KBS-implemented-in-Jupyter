```python
#simple example of a basic KBS

class KnowledgeBase:
    def __init__(self): # Initialize the knowledge base
        self.facts = set()
        self.rules = []

    def add_fact(self, fact):# Add a fact to the knowledge base
        self.facts.add(fact)

    def add_rule(self, rule):# Add a rule to the knowledge base
        self.rules.append(rule)

    def infer(self):# Infer new facts based on existing facts and rules
        new_facts = set()# Iterate through rules and apply them
        for rule in self.rules:
            if rule['condition'](self.facts):
                new_facts.add(rule['conclusion'])
        self.facts.update(new_facts)
        return new_facts# Example usage
kb = KnowledgeBase()# Create a knowledge base instance
kb.add_fact('It is raining')# Add a fact
kb.add_rule({
'condition': lambda facts: 'It is raining' in facts, # Rule condition
    'conclusion': 'The ground is wet'
})
new_inferences = kb.infer()# Infer new facts
print("Newly inferred facts:", new_inferences)
print("All facts in the knowledge base:", kb.facts)

kb.add_fact('The ground is wet')# Add another fact
kb.add_rule({'condition': lambda facts: 'The ground is wet' in facts, # Another rule condition
    'conclusion': 'The grass is green'
})
new_inferences = kb.infer()# Infer new facts
print("Newly inferred facts:", new_inferences)
print("All facts in the knowledge base:", kb.facts)


#!/usr/bin/env python3
```
